> Number of samples per class

```python
unique, counts = np.unique(ytrain, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
# Displaying a few Training Samples
labels = ['A','B','C','D','E','F','G','H','I']

for j in range(len(labels)):
    idx = list(np.where(ytrain == labels[j])[0])
    random.shuffle(idx)
    # Only displaying 10 random samples for each character
    for k in range(10):
        plt.imshow(Xtrain[idx[k],:,:,:]);
        plt.title(ytrain[idx[k]][0])
        plt.show();
```

## Conclusion
Overall, the quality of the training set is pretty good. There are 3205 training samples per class in the given training data set which is a large number and probably large enough to make accurate predictions. 

The dimensionality of the feature space is 100 x 100 x 3 meaning there are 30,000 features in the training data (representing each pixel's color channel). The dimensionality is incredibly high and generally unusable without using feature extraction. Without the feature extraction, the data suffers from the curse of dimensionality. 

The most important part that makes the data unreliable is the representation and variability in the data. The majority of the images are in the same alignment with the palm facing forward and the sign facing upwards but there are a number that are facing sideways or other directions. The backgrounds are rather inconsistent with different textures and colors that could incidentally create false edges. Some backgrounds have other objects such as keyboards and mice that could create false edges that the models may recognize. There are several images that have a different signing hand as they use their left hand instead of right hand. Another point of variation comes in the hand shape people made, some people kept fingers together while others had their fingers spread apart.
